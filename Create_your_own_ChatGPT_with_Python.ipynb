```python
from google.colab import userdata
from openai import OpenAI
```

```python
class ChatSession:
    def __init__(self, model="gpt-3.5-turbo"):
        self.client = OpenAI(api_key=userdata.get('OPENAI_API_KEY'))
        self.model = model
        # 最初にシステムの役割を持つメッセージを初期化
        self.messages = [{"role": "system", "content": "You are a helpful assistant."}]

    def ask(self, user_input):
        # ユーザーからの入力をメッセージリストに追加
        self.messages.append({"role": "user", "content": user_input})

        # モデルによる応答を生成
        chat_completion = self.client.chat.completions.create(
            model=self.model,
            messages=self.messages,
            max_tokens=1024,
            n=1,
            stop=None,
            temperature=0.5,
        )

        # 応答をメッセージリストに追加して会話を更新
        response = chat_completion.choices[0].message.content
        self.messages.append({"role": "user", "content": response})

        return response
```

```python
session = ChatSession()  # 新しいチャットセッションを生成
```

```python
prompt = """
My name is Tomoya.
"""
session.ask(prompt)
```

```python
prompt = """
What's my name?
"""
session.ask(prompt)
```
