```python
right_df_en.columns
```

```python
df_en = pd.concat([left_df_en, right_df_en], axis=0)
df_en
```

```python
df_en.to_csv("./data/mem.csv")
```

```python
import re

# Function to convert time in "hh:mm:ss" format to seconds
def time_to_seconds(time_str):
    parts = list(map(int, re.findall(r'\d+', time_str)))
    if len(parts) == 3:
        return parts[0] * 3600 + parts[1] * 60 + parts[2]
    elif len(parts) == 2:
        return parts[0] * 60 + parts[1]
    else:
        return parts[0]

# Convert all time and pace columns to seconds
time_columns = ['5km_Time_Trial', 'Pace_per_km_5km', '10km_Time_Trial', 'Pace_per_km_10km_Half_Marathon',
                'Expected_Half_Marathon_Time', 'Pace_per_km_Half', 'Expected_Full_Marathon_Time', 'Pace_per_km_Full']

for column in time_columns:
    df_en[column] = df_en[column].apply(time_to_seconds)

df_en.head()
```

```python
df_en_2 = df_en.drop(['Pace_per_km_5km', 'Pace_per_km_10km_Half_Marathon', 'Pace_per_km_Half', 'Pace_per_km_Full'], axis=1)
```

```python
df_en_2['Grade'] = df_en_2['Grade'].str.replace('級', '')
```

```python
df_en_2['Grade'] = pd.to_numeric(df_en_2['Grade'], errors='coerce')
```
